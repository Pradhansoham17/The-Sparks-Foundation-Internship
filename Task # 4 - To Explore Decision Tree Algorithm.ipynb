<font size='5'><b>Importing the Libraries :-</b></font>

```python
# after removing data row as index row and changing column names
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

<font size='5'><b>Loading the Iris Dataset into a Pandas Dataframe :-</b></font>

```python
df = pd.read_csv('C:\\Users\\soham\\Downloads\\Iris.csv')
df = df.drop(['Id'], axis=1)
df  
```

```python
sns.pairplot(df, hue="Species")
```

<font size='5'><b>Splitting the Dataset into training and testing :-</b></font>

```python
from sklearn.model_selection import train_test_split
X, y = df.iloc[:,:-1],df.iloc[:,[-1]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
X_train
```

```python
X_test
```

<font size='5'><b>Applying the Sklearn Decision-Tree Classifier on our training data :-</b></font>

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
X = X_train
y = y_train
clf.fit(X,y)
```

```python
clf.classes_
```

```python
clf.n_features_
```

<font size='5'><b>Predicting the results of Test Data and Checking Accuracy score :-</b></font>

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
clf.score(X_test,y_test)
```

<font size='5'><b>Confusion Matrix and Classification Report :-</b></font>

```python
from sklearn.metrics import confusion_matrix
y_true = y_test['Species']
confusion_matrix(y_true, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred, target_names=clf.classes_))
```

<font size='5'><b>Visualizing the Decision Tree formed by our model :-</b></font>

```python
from sklearn import tree
fig, ax = plt.subplots(figsize=(12, 12))
tree.plot_tree(clf,feature_names=df.iloc[:,:-1].columns,  
                                class_names =clf.classes_, filled=True, rounded=True, fontsize=12)
plt.show()
```

<font size='5'><b>Using Graphviz to export our Decision Tree to a PDF File :-</b></font>

```python
import graphviz 
dot_data = tree.export_graphviz(clf,feature_names=df.iloc[:,:-1].columns,  
                                class_names = clf.classes_, filled=True, rounded=True,  
                                special_characters=True, out_file=None) 

graph = graphviz.Source(dot_data) 
graph.render("Graphviz_Decision Tree") 
```
